```python
import torch
print("PyTorch 버전:", torch.__version__)

# GPU가 있으면 cuda, 없으면 cpu를 자동 선택 (앞으로 모든 실습 공통 패턴)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("사용 장치:", device)
```
