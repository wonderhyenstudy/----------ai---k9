```python
import torch
x = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])
print(x)
print("shape:", x.shape)   # torch.Size([2, 3])
print("dtype:", x.dtype)
```

```python
w = torch.tensor(2.0, requires_grad=True)   # 기울기 추적 켜기
y = w ** 2
z = 2 * y + 5         # z = 2w^2 + 5
z.backward()          # dz/dw 자동 계산
print(w.grad)         # dz/dw = 4w = 4*2.0 = 8.0
```

```python
import torch, torch.nn as nn, torch.optim as optim
model = nn.Linear(1, 1)                     # y = wx + b
criterion = nn.MSELoss()                    # 손실 함수
optimizer = optim.SGD(model.parameters(), lr=0.01)

X = torch.tensor([[1.0], [2.0], [3.0]])
y = torch.tensor([[2.0], [4.0], [6.0]])     # y = 2x

for epoch in range(100):
    optimizer.zero_grad()        # 기울기 초기화
    pred = model(X)              # 순전파(예측)
    loss = criterion(pred, y)    # 손실 계산
    loss.backward()              # 역전파
    optimizer.step()             # 가중치 갱신
print("학습된 가중치:", model.weight.item())
```
