```python
def add(a, b):       # a, b는 매개변수
    return a + b     # 결과 반환

result = add(3, 5)   # 3, 5는 인수
print(result)        # 8
```

```python
def greet(name, msg="안녕하세요"):
    print(f"{name}님, {msg}")

greet("철수")              # 철수님, 안녕하세요
greet("영희", "반갑습니다")  # 영희님, 반갑습니다
```

```python
def total(*nums):     # nums는 튜플로 묶임
    return sum(nums)

print(total(1, 2, 3))        # 6
print(total(10, 20, 30, 40)) # 100
```

```python
def profile(**info):   # info는 딕셔너리로 묶임
    for key, value in info.items():
        print(f"{key}: {value}")

profile(이름="이상용", 직업="강사", 지역="부산", 오늘점심뭐먹죠="라면")
```

```python
def change(items):
    items.append(99)    # 원본 리스트가 바뀜! (가변)

my_list = [1, 2, 3]
change(my_list)
change(my_list)
print(my_list)          # [1, 2, 3, 99]
```

```python
# 문자열 슬라이싱 (ch3 예제)
s = "abc"
print(s[0:2])   # ab  (인덱스 0~1)
print(s[::-1])  # cba (거꿁로)

# lambda 예시
double = lambda x: x * 2
print(double(7))  # 14
```
