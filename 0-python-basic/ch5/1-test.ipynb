```python
class Animal:
    def __init__(self, name, age):   # 생성자: 객체 만들 때 실행
        self.name = name             # 속성 저장
        self.age = age
    def speak(self):                 # 메서드
        print("말하기")

lsy = Animal("lsy", 20)   # 객체(인스턴스) 생성
lsy.speak()               # 말하기
print(lsy.name, lsy.age)  # lsy 20
```

```python
class Dog(Animal):                   # Animal을 상속
    def __init__(self, name, age, food):
        super().__init__(name, age)  # 부모 생성자 호출
        self.food = food
    def speak(self):                 # 부모 speak() 재정의
        print("멍멍")
    def show_food(self):
        print(f"좋아하는 음식: {self.food}")

d = Dog("끜돌이", 3, "소세지")
d.speak()       # 멍멍
d.show_food()   # 좋아하는 음식: 소세지
```

```python
class Account:
    def __init__(self, owner):
        self.owner = owner
        self.__balance = 0          # 외부에서 직접 접근 차단
    def deposit(self, amount):
        self.__balance += amount
    def get_balance(self):          # 조회는 메서드로
        return self.__balance

a = Account("이상용")
a.deposit(1000)
print(a.get_balance())   # 1000
# print(a.__balance)     # 에러! 직접 접근 불가
```

```python
class Book:
    def __init__(self, title, pages):
        self.title = title
        self.pages = pages
    def __str__(self):
        return f"《{self.title}》 ({self.pages}p)"

b = Book("파이썬 기본서", 300)
print(b)   # 《파이썬 기본서》 (300p)
```
