```python
# 기존 for문 방식
squares = []
for x in range(1, 6):
    squares.append(x ** 2)
print(squares)   # [1, 4, 9, 16, 25]

# 컴프리헨션 방식 (한 줄!)
squares = [x ** 2 for x in range(1, 6)]
print(squares)   # [1, 4, 9, 16, 25]

# 조건 붙이기
evens = [x for x in range(10) if x % 2 == 0]
print(evens)     # [0, 2, 4, 6, 8]
```

```python
fruits = ["사과", "바나나", "체리"]
for index, fruit in enumerate(fruits):
    print(f"인덱스 {index}: {fruit}")
# 인덱스 0: 사과 / 인덱스 1: 바나나 / 인덱스 2: 체리

# 시작 번호 바꾸기
for no, fruit in enumerate(fruits, start=1):
    print(no, fruit)   # 1 사과 ...
```

```python
names = ["철수", "영희", "민수"]
scores = [85, 90, 78]
for name, score in zip(names, scores):
    print(f"{name}의 점수: {score}")
```

```python
# 일반 함수
def add(a, b):
    return a + b
print(add(3, 5))   # 8

# 람다로 동일하게
add_lambda = lambda a, b: a + b
print(add_lambda(3, 5))   # 8

# 정렬 기준으로 자주 사용
people = [("철수", 25), ("영희", 22)]
print(sorted(people, key=lambda p: p[1]))  # 나이순 정렬
```

```python
def sum_numbers(*args):       # 개수 제한 없이 → 튜플
    return sum(args)
print(sum_numbers(1, 2, 3))         # 6
print(sum_numbers(10, 20, 30, 40))  # 100

def print_user(**kwargs):     # 키워드 여러 개 → 딕셔너리
    for key, value in kwargs.items():
        print(f"{key}: {value}")
print_user(name="철수", age=25, city="서울")
```

```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]
merged = [*list1, *list2]
print(merged)        # [1, 2, 3, 4, 5, 6]

def add(a, b, c):
    return a + b + c
numbers = [3, 4, 5]
print(add(*numbers)) # 12 (리스트를 낱개 인수로 풀어서 전달)
```
